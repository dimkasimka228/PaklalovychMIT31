Крок 0: Імпорт необхідних бібліотек
Спочатку імпортуємо всі бібліотеки, які нам знадобляться для генерації даних, їх обробки, тренування моделей та оцінки.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Генератори даних
from sklearn.datasets import make_classification

# Обробка даних
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer # Для обробки пропущених значень

# Модель
from sklearn.neighbors import KNeighborsClassifier # Візьмемо KNN як приклад моделі

# Метрики
from sklearn.metrics import (
    accuracy_score, 
    precision_score, 
    recall_score, 
    f1_score, 
    classification_report, 
    confusion_matrix
)

# Налаштування візуалізацій
%matplotlib inline
sns.set(style="whitegrid")
```

<h3>Частина 1. Згенерований набір даних (DatasetGenerator)
Створимо штучний набір даних для бінарної класифікації за допомогою make_classification.

```python
# Cтворюємо набір даних: 1000 зразків, 10 ознак (з них 8 інформативних), 2 кластери на клас
X_gen, y_gen = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=8,
    n_redundant=2,
    n_clusters_per_class=2,
    class_sep=1.0,
    random_state=42
)

# Перетворимо у DataFrame для зручності
features_gen = [f'feature_{i}' for i in range(X_gen.shape[1])]
df_gen = pd.DataFrame(X_gen, columns=features_gen)
df_gen['target'] = y_gen

print("Перші 5 рядків згенерованих даних:")
print(df_gen.head())

print("\nРозподіл класів:")
print(df_gen['target'].value_counts(normalize=True))
```

1.2. Кроки моделювання
Виконаємо всі необхідні кроки попередньої обробки .

```python
# Визначимо X та y
X = df_gen.drop('target', axis=1)
y = df_gen['target']

# 1. Обробка пропущених значень 
# make_classification не створює пропуски, але перевірка є гарною практикою
print(f"\nПропущені значення: {X.isnull().sum().sum()}") 
# (Якщо б вони були, ми б використали SimpleImputer)

# 2. Кодування категоріальних змінних 
# Всі дані вже числові, кодування не потрібне.

# 3. Розподіл на навчальну та тестову вибірки 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 4. Масштабування (Стандартизація) 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

1.3. Навчання та оцінка моделі

```python
# Ініціалізація моделі
model_gen = KNeighborsClassifier(n_neighbors=5)

# Навчання моделі
model_gen.fit(X_train_scaled, y_train)

# Прогнозування
y_pred_gen = model_gen.predict(X_test_scaled)

# Оцінка моделі
print("--- Оцінка моделі на згенерованих даних ---")
print(classification_report(y_test, y_pred_gen))

print(f"Accuracy: {accuracy_score(y_test, y_pred_gen):.4f}")
print(f"Precision: {precision_score(y_test, y_pred_gen):.4f}")
print(f"Recall: {recall_score(y_test, y_pred_gen):.4f}")
print(f"F1-score: {f1_score(y_test, y_pred_gen):.4f} ")
```

<h3>Частина 2. Набір даних з Kaggle: "Red Wine Quality"

2.1. Завантаження та аналіз даних

```python
# Завантаження даних
try:
    df_wine = pd.read_csv('winequality-red.csv')
except FileNotFoundError:
    print("ПОМИЛКА: Файл 'winequality-red.csv' не знайдено.")
    print("Будь ласка, завантаж його з Kaggle і поклади в ту ж папку, що і ноутбук.")
    # Зупинимо виконання цієї частини, якщо файлу немає
    df_wine = None

if df_wine is not None:
    print("Перші 5 рядків даних 'Red Wine Quality':")
    print(df_wine.head())

    print("\nІнформація про датасет:")
    df_wine.info()

    print("\nОписова статистика:")
    print(df_wine.describe())

    print("\nРозподіл класів 'quality':")
    print(df_wine['quality'].value_counts())
```

2.2. Кроки моделювання



2.2.1. Попередня обробка даних

```python
if df_wine is not None:
    # 1. Обробка пропущених значень 
    print(f"\nПропущені значення: {df_wine.isnull().sum().sum()}")
    # (Цей датасет відомий тим, що не має пропусків)

    # 2. Інженерія ознак (перетворення на бінарну класифікацію) 
    # Оцінки 'quality' (від 3 до 8) є порядковими.
    # Перетворимо це на задачу бінарної класифікації: "погане" (0) vs "гарне" (1) вино.
    # Вважатимемо вино з оцінкою > 5 "гарним" (1), а <= 5 "поганим" (0).
    df_wine['target'] = df_wine['quality'].apply(lambda q: 1 if q > 5 else 0)
    
    # Видаляємо оригінальну колонку 'quality'
    df_wine = df_wine.drop('quality', axis=1)

    print("\nРозподіл нових бінарних класів:")
    print(df_wine['target'].value_counts(normalize=True))
    # Як видно, класи дещо незбалансовані 

    # Визначимо X та y
    X = df_wine.drop('target', axis=1)
    y = df_wine['target']

    # 3. Кодування категоріальних змінних 
    # Всі дані вже числові, кодування не потрібне.

    # 4. Розподіл на навчальну та тестову вибірки 
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # stratify=y для збалансованого розподілу

    # 5. Масштабування (Стандартизація) 
    # Це важливо, оскільки ознаки (наприклад, 'free sulfur dioxide' і 'pH') мають різний масштаб
    scaler_wine = StandardScaler()
    X_train_scaled = scaler_wine.fit_transform(X_train)
    X_test_scaled = scaler_wine.transform(X_test)
```

2.2.2. Навчання та оцінка моделі

```python
if df_wine is not None:
    # Ініціалізація моделі
    model_wine = KNeighborsClassifier(n_neighbors=5)

    # Навчання моделі
    model_wine.fit(X_train_scaled, y_train)

    # Прогнозування
    y_pred_wine = model_wine.predict(X_test_scaled)

    # Оцінка моделі
    print("--- Оцінка моделі на даних 'Red Wine Quality' ---")
    print(classification_report(y_test, y_pred_wine))

    # Візуалізація матриці помилок
    cm = confusion_matrix(y_test, y_pred_wine)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Погане (0)', 'Гарне (1)'], yticklabels=['Погане (0)', 'Гарне (1)'])
    plt.xlabel('Передбачено')
    plt.ylabel('Реально')
    plt.title('Матриця помилок')
    plt.show()
```
