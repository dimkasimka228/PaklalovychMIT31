## Імпорт бібліотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def log(message, variable):
    print(f"{message}: {variable}")

def visualize(x, y, y_pred):
    plt.scatter(X, y, color='blue', label='Фактичні дані')
    plt.plot(X, y_pred, color='red', linewidth=2, label='Лінія регресії')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Лінійна регресія')
    plt.legend()
    plt.show()
```

## Підготовка днних

```python
np.random.seed(100)
# 100 чисел вiд 0 до 10
X = np.linspace(0, 10, 100)

true_slope = 2
true_intercept = 2
# нахил (наскільки круто йде лінія)               ; лінія росте вдвічі швидше, ніж X
# перетин з віссю Y (де лінія перетинає Y при X=0); при X=0, Y=2

y = true_slope * X + true_intercept + np.random.normal(0, 1, 100)
# 100 випадкових чисел (Середнє = 0; Стандартне відхилення = 1; Тобто: більшість чисел від -3 до +3, але можуть бути й рідкісні викиди)
```

##  Створення моделі

```python
# Creating model
model = LinearRegression()

# Teaching model
model.fit(X.reshape(-1, 1), y)
```

## Прогнозовані значення

```python
y_pred = model.predict(X.reshape(-1, 1))
```

## Оцінка моделі

```python
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

log("Середньоквадратична похибка (MSE)", mse)
log("Середньо абсолютна похибка (MAE)", mae)
log("Коефіцієнт детермінації (R^2)", r2)
```

## Результат

```python
visualize(X, y, y_pred)
```
