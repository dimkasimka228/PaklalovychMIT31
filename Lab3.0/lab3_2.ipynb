## Імпорт бібліотек

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def log(message, variable):
    print(f"{message}: {variable}")

def visualize(x, y, y_pred):
    plt.scatter(X, y, color='blue', label='Фактичні дані')
    plt.plot(X, y_pred, color='red', linewidth=2, label='Лінія регресії')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Лінійна регресія')
    plt.legend()
    plt.show()
```

## Підготовка данних

```python
df = pd.read_csv('simple_linear_regression.csv')

X = df[['SAT']]
y = df['GPA']
```

## Створення моделі

```python
# Creating model
model = LinearRegression()

# Teaching model
model.fit(X, y)
```

## Прогнозовані значення

```python
y_pred = model.predict(X)
```

## Оцінка моделі

```python
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

log("Середньоквадратична похибка (MSE)", mse)
log("Середньо абсолютна похибка (MAE)", mae)
log("Коефіцієнт детермінації (R^2)", r2)
```

## Результат

```python
visualize(X, y, y_pred)
```
