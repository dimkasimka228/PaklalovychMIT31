```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Завантаження даних
df = pd.read_csv("Top_20_Pakistani_Universities.csv")

# 2. Перевірка структури
print("Колонки:", df.columns.tolist())
print(df.head())

# 3. Кількість університетів за типом
df['University Type'].value_counts().plot(kind='bar', color='teal', figsize=(8,4))
plt.title("Кількість університетів за типом")
plt.xlabel("Тип університету")
plt.ylabel("Кількість")
plt.tight_layout()
plt.show()

# 4. Перетворення QS Ranking (Asia) у числовий формат
def parse_rank(value):
    try:
        if isinstance(value, str) and "-" in value:
            parts = value.split("-")
            return (int(parts[0]) + int(parts[1])) / 2   # середнє значення діапазону
        else:
            return float(value)
    except:
        return None

df['QS_Ranking_Asia_Numeric'] = df['QS Ranking (Asia)'].apply(parse_rank)

# 5. Середній QS Ranking (Asia) за типом університету
df.groupby('University Type')['QS_Ranking_Asia_Numeric'].mean().plot(kind='bar', color='orange', figsize=(8,4))
plt.title("Середній QS Ranking (Asia) за типом")
plt.ylabel("Середній рейтинг (менше = краще)")
plt.tight_layout()
plt.show()

# 6. Топ-10 університетів за QS Ranking (Asia)
top10 = df[['University Name','QS Ranking (Asia)','Location','University Type','QS_Ranking_Asia_Numeric']] \
          .sort_values('QS_Ranking_Asia_Numeric') \
          .head(10)

print("\nТоп-10 університетів за QS Ranking (Asia):")
print(top10)
```
