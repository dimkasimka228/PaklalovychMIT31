```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def log(message, variable):
    print(f"{message}: {variable}")

def visualize(x, y, y_pred):
    plt.scatter(X, y, color='blue', label='Фактичні дані')
    plt.plot(X, y_pred, color='red', linewidth=2, label='Лінія регресії')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.title('Лінійна регресія')
    plt.legend()
    plt.show()
```

```python
from sklearn.datasets import make_regression
```

## Формування DataFrame
підготовка для моделювання


```python
_X, _y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

df = pd.DataFrame( _X, columns=['X'])
df['y'] = _y

X = df[['X']]
y = df['y']
```

## Створення моделі

```python
# Creating model
model = LinearRegression()

# Teaching model
model.fit(X, y)
```

## Передбачення значень

```python
y_pred = model.predict(X)
```

## Оцінка моделей

```python
mse = mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

log("Середньоквадратична похибка (MSE)", mse)
log("Середньо абсолютна похибка (MAE)", mae)
log("Коефіцієнт детермінації (R^2)", r2)
```

## результат

```python
visualize(X, y, y_pred)
```
