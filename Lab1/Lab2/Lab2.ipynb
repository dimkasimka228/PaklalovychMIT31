```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Лінійна функція: Y = 2X + 3
X = np.arange(0, 10, 0.5)
Y = 2 * X + 3

# Додавання шуму
noise = np.random.normal(0, 2, len(X))
Y_hat = Y + noise

# Побудова графіка
plt.plot(X, Y, label='True Function', color='blue')
plt.scatter(X, Y_hat, label='Measured Data', color='red')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Лінійна функція з шумом')
plt.grid(True)
plt.show()

# Обчислення похибок
mAE = mean_absolute_error(Y, Y_hat)
mSE = mean_squared_error(Y, Y_hat)

# Запис у CSV
results = pd.DataFrame({'X': X, 'Y': Y, 'Y_hat': Y_hat})
results['mAE'] = mAE
results['mSE'] = mSE
results.to_csv('results.csv', index=False)
```
