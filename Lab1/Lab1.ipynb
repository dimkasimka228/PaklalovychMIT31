```python
import numpy as np

# Створення масиву з 10 випадкових чисел
random_array = np.random.rand(10)

# Виведення масиву
print("Масив:", random_array)

# Обчислення середнього значення
average = np.mean(random_array)
print("Середнє значення:", average)
```
